# <b><div style='padding:15px;background-color:#9264E3;color:white;border-radius:2px;font-size:110%;text-align: center'>Divide a data em meses e anos </div></b>

```python
df['Initial release date'] = pd.to_datetime(df['Initial release date'])
df['Year'] = df['Initial release date'].dt.year
df['Month'] = df['Initial release date'].dt.month
```

# <b><div style='padding:15px;background-color:#9264E3;color:white;border-radius:2px;font-size:110%;text-align: center'>As 20 maiores vendas de todos os tempos </div></b>

```python
top_20 = df.nlargest(20, 'Sales')
```

```python
import math

millnames = ['',' Thousand',' Million',' Billion',' Trillion']

def millify(n):
    n = float(n)
    millidx = max(0,min(len(millnames)-1,
                        int(math.floor(0 if n == 0 else math.log10(abs(n))/3))))

    return '{:.0f}{}'.format(n / 10**(3 * millidx), millnames[millidx])
```

```python
values = []
for x in top_20.Sales:
    values.append(millify(x))
```

```python
fig, ax = plt.subplots(figsize=(15, 15))
palette = sns.color_palette("Purples",n_colors=24)
palette.reverse()
ax = sns.barplot(ax =ax,data = top_20 , y = top_20['Title'], x = top_20['Sales'],palette = palette)
ax.set(ylabel='Games',xlabel='Sales', title='Top Sales of All Time')

ax.set_yticklabels( ax.get_yticklabels(), fontsize = 16)

ax.bar_label(ax.containers[0], labels=values, padding=3, fontsize = 18)
```

# <b><div style='padding:15px;background-color:#9264E3;color:white;border-radius:2px;font-size:110%;text-align: center'>10 principais editores </div></b>

```python
fig, ax = plt.subplots()
palette = sns.color_palette("Greens",n_colors=12)
palette.reverse()
ax = sns.barplot(ax =ax,data = top_20.head(10) , y = top_20['Publisher(s)'], x = top_20['Sales'],palette = palette)
ax.set(ylabel='Publisher(s)',xlabel='Sales', title='Top 10 Publisher(s)')
```
